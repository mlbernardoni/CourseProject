```python
import sys
from datetime import datetime
import matplotlib.pyplot as plt

%matplotlib notebook
#print(sys.path)
#sys.path.append('..')


from plsa import Corpus, Pipeline, Visualize

from plsa.pipeline import DEFAULT_PIPELINE
from plsa.algorithms import PLSA
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)
directory = 'psla_data/PSLAreduced.csv'
#directory = 'testdata'

pipeline = Pipeline(*DEFAULT_PIPELINE)

corpus = Corpus.from_csv(directory, pipeline, -1, 'latin_1', 2700)

n_topics = 30
plsa = PLSA(corpus, n_topics, True)
result = plsa.fit()

result = plsa.best_of(5)
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)
```

```python
visualize = Visualize(result)
```

```python
fig, ax = plt.subplots()
_ = visualize.topics(ax)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(9.4, 10))
_ = visualize.wordclouds(fig)
```

```python
fig, ax = plt.subplots()
_ = visualize.topics_in_doc(50, ax)
fig.tight_layout()
```
